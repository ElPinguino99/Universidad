```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del problema
U0 = 1.0   # Joules
a  = 2.0   # metros

# Definimos un rango de x > 0 (sin llegar a cero para evitar divergencia)
x = np.linspace(0.1, 10, 1000)  # desde 0.1 hasta 10 m

# Definimos el potencial U(x)
U = U0 * (a/x + x/a)

# Graficamos
plt.figure(figsize=(7,5))
plt.plot(x, U, label=r'$U(x) = U_0\left(\frac{a}{x} + \frac{x}{a}\right)$')
plt.title('Problema 12: Potencial $U(x)$')
plt.xlabel('x (m)')
plt.ylabel('U(x) (J)')
plt.ylim(0, 10)   # Ajusta según la escala que desees ver
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del problema
m = 2.0        # kg
A = 10.0       # N
B = 2.0        # N/s
x0 = 5.0       # m
v0 = 0.0       # m/s

# Definimos un rango de tiempo
t = np.linspace(0, 10, 200)  # 0 a 10 s

# Fuerza en función del tiempo
F = A - B*t

# Velocidad en función del tiempo
v = v0 + (A/m)*t - (B/(2*m))*(t**2)

# Posición en función del tiempo
x = x0 + v0*t + (A/(2*m))*(t**2) - (B/(6*m))*(t**3)

# Graficamos en subplots
fig, axs = plt.subplots(3, 1, figsize=(7,10), sharex=True)

# 1) Gráfica de la Fuerza
axs[0].plot(t, F, 'r-', label=r'$F(t) = A - Bt$')
axs[0].set_ylabel('Fuerza (N)')
axs[0].set_title('Problema 14')
axs[0].grid(True)
axs[0].legend()

# 2) Gráfica de la Velocidad
axs[1].plot(t, v, 'b-', label=r'$v(t)$')
axs[1].set_ylabel('Velocidad (m/s)')
axs[1].grid(True)
axs[1].legend()

# 3) Gráfica de la Posición
axs[2].plot(t, x, 'g-', label=r'$x(t)$')
axs[2].set_xlabel('Tiempo (s)')
axs[2].set_ylabel('Posición (m)')
axs[2].grid(True)
axs[2].legend()

plt.tight_layout()
plt.show()
```
